```python
all = soup.find_all("div",{"class": "list-card-info"})
len(all)
all[0].find("div",{"class": "list-card-price"}).text.replace("\n","")
all[0].find("address",{"class": "list-card-addr"}).text.replace("\n","")
all[0].find("ul",{"class": "list-card-details"}).text.replace("\n","")
all[0].find("a", {"class":"list-card-link list-card-link-top-margin"}).get("href")

try: 
  pages=soup.find_all("li",{"aria-current": "page"})[-1].text
except:
  pages=None
page_num = pages[-2:]
```

```python
List=[]
base_url_1="https://www.zillow.com/seattle-wa/2-_beds/2.0-_baths"
base_url_2="_p/?searchQueryState=%7B%22pagination%22%3A%7B%22currentPage%22%3A3%7D%2C%22usersSearchTerm%22%3A%22Seattle%2C%20WA%22%2C%22mapBounds%22%3A%7B%22west%22%3A-125.67605618115235%2C%22east%22%3A-118.60086086865235%2C%22south%22%3A45.52685943727079%2C%22north%22%3A49.55133503508338%7D%2C%22mapZoom%22%3A7%2C%22regionSelection%22%3A%5B%7B%22regionId%22%3A16037%2C%22regionType%22%3A6%7D%5D%2C%22isMapVisible%22%3Atrue%2C%22filterState%22%3A%7B%22beds%22%3A%7B%22min%22%3A2%7D%2C%22baths%22%3A%7B%22min%22%3A2%7D%2C%22ah%22%3A%7B%22value%22%3Atrue%7D%2C%22sort%22%3A%7B%22value%22%3A%22globalrelevanceex%22%7D%7D%2C%22isListVisible%22%3Atrue%7D"

for page in range(0,int(page_num),1):
  print(base_url_1 + str(page) + base_url_2)
  c=res.content
  soup=BeautifulSoup(c,"html.parser")
  all=soup.find_all("div",{"class":"list-card-info"})
  for item in all:
    d={}
    try:  
      d["Price"]=item.find("div",{"class": "list-card-price"}).text
    except:
      d["Price"]=None
    try:
      d["Address"]=item.find("address",{"class": "list-card-addr"}).text.replace("\n","")
    except:
      d["Address"]=None
    try:
      d["Details"]=item.find("ul",{"class": "list-card-details"}).text
    except:
      d["Details"]=None
    try:
      d["Link"]=item.find("a", {"class":"list-card-link list-card-link-top-margin"}).get("href")
    except:
      d["Link"]=None
    
    List.append(d)

  
```
